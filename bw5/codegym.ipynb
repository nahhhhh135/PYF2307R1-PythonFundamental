```python
import turtle
p=turtle.Turtle()
p.speed(30)
p.color('#000000')
p.fillcolor("#FFFF00")

p.getscreen().bgcolor("#00FF00")
#gọi hàm vẽ hình tam giác
def draw_triangle(side,fillcolor = '#FFFF00' ,color = '#000000'):
    p.color(color)
    p.fillcolor(fillcolor)
    p.begin_fill()
    p.circle(side,steps = 3)
    p.end_fill()

p.left(60)
draw_triangle(100)

def draw_rectangular(h,w,fillcolor = '#FFFF00' ,color = '#000000'):
    p.color(color)
    p.fillcolor(fillcolor)
    p.begin_fill()
    for i in range(2):
        p.forward(h)
        p.right(90)
        p.forward(w)
        p.right(90)
    p.end_fill()
    
    
p.right(60)
p.goto(-180,0)
draw_rectangular(300,150,"#97D433",'#2E6B95')

p.penup()
p.goto(-130,-50)
p.pendown()
draw_rectangular(50,100,"#97D433",'#2E6B95')
p.hideturtle()
p.penup()
p.goto(-250,0)
p.pendown()
p.hideturtle()
p.left(60)
draw_triangle(30,"#97D433",'#2E6B95')
p.right(60)

p.penup()
p.goto(-240,-60)
p.pendown()

p.left(60)
draw_triangle(40,"#33D447",'#2E6B95')
p.right(60)

p.penup()
p.goto(-225,-150)
p.pendown()

p.left(60)
draw_triangle(60,"#33D447",'#2E6B95')
p.right(60)

p.penup()
p.backward(30)
p.right(90)
p.pendown()
draw_rectangular(20,30,"#95442E",'#2E6B95')
p.hideturtle()

#vẽ cây thông còn lại

p.penup()
p.goto(320,0)
p.pendown()
p.left(150)
draw_triangle(40,"#D46233",'#2E6B95')

p.penup()
p.goto(340,-90)
p.pendown()
p.right(60)
p.left(60)
draw_triangle(60,"#D46233",'#2E6B95')
p.right(60)

p.penup()
p.goto(360,-210)
p.pendown()
p.left(60)
draw_triangle(80,"#D46233",'#2E6B95')
p.right(60)

p.penup()
p.backward(50)
p.right(90)
p.pendown()
draw_rectangular(30,40,"#95442E",'#2E6B95')

p.penup()
p.goto(350,400)
p.pendown()
p.color('#000000')
p.fillcolor("#FFFF00")
p.begin_fill()
p.circle(200)
p.end_fill()


p.hideturtle()

turtle.done()
```
