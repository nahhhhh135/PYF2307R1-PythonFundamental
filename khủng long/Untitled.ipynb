```python
import tkinter as tk
from PIL import Image, ImageTk
import random
# Tạo cửa sổ game
root = tk.Tk()
root.title("Game khủng long")
root.geometry("600x400")
# Tải hình ảnh khủng long
dragon_image = Image.open("dragon.jng")
dragon = ImageTk.PhotoImage(dragon_image)

# Tạo một danh sách các chướng ngại vật
obstacles = []
for i in range(10):
    obstacle_image = Image.open("obstacle.jng")
    obstacle = ImageTk.PhotoImage(obstacle_image)
    obstacles.append(obstacle)
# Vị trí hiện tại của khủng long
dinosaur_x = 200
dinosaur_y = 200
# Hàm di chuyển khủng long
def move_dinosaur(event):
    global dinosaur_x
    if event.keysym == "Up":
        dragon_y -= 10
    elif event.keysym == "Down":
        dragon_y += 10
# Hàm tạo chướng ngại vật mới
def create_obstacle():
    global obstacles
    obstacle = obstacles.pop()
    obstacle_x = random.randint(0, 500)
    obstacle_y = 0
    return obstacle, obstacle_x, obstacle_y
# Hàm kiểm tra va chạm giữa khủng long và chướng ngại vật
def check_collision(dinosaur, obstacle):
    if dragon.x <= obstacle.x + obstacle.width and dragon.x + dragon.width >= obstacle.x and dragon.y <= obstacle.y + obstacle.height and dragon.y + dragon.height >= obstacle.y:
        return True
    return False
# Hàm vẽ các đối tượng trên màn hình
def draw_objects():
    # Vẽ khủng long
    canvas.create_image(dinosaur_x, dragon_y, image=dragon)
    # Vẽ chướng ngại vật
    for obstacle in obstacles:
        canvas.create_image(obstacle_x, obstacle_y, image=obstacle)
# Vòng lặp game
while True:
    # Cập nhật vị trí của khủng long
    dragon_x += 10

    # Tạo chướng ngại vật mới
    obstacle, obstacle_x, obstacle_y = create_obstacle()

    # Kiểm tra va chạm
    if check_collision(dragon, obstacle):
        # Game over
        break

    # Vẽ các đối tượng trên màn hình
    draw_objects()

    # Cập nhật giao diện
    root.update()

    # Tạo canvas
    canvas = tk.Canvas(root)
    canvas.pack()

    # Gắn sự kiện bàn phím
    root.bind("<KeyPress>", move_dragon)

    # Vẽ các đối tượng trên màn hình
    draw_objects()

    # Chạy vòng lặp game
    root.mainloop()
```
